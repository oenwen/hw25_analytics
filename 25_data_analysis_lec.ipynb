```python
df1 = pd.read_csv('names/yob1880.txt', names = ['Name', 'Gender', 'Quantity'])
df1['Year'] = 1880
df2 = pd.read_csv('names/yob2017.txt', names = ['Name', 'Gender', 'Quantity'])
df2['Year'] = 2017
df3 = pd.read_csv('names/yob1950.txt', names = ['Name', 'Gender', 'Quantity'])
df3['Year'] = 1950
df_res = pd.concat([df1, df2, df3])
df_res
```

```python
df_res.reset_index(inplace = True)
df_res
```

```python
df_res[df_res['Name'] == 'Arthur']
```

```python
df_res.groupby(['Name', 'Gender']).sum()['Quantity']
df_res
```

Создадим в датафрейме отдельный столбец "Длина имени"
    

```python
df_res['Name_length'] = df_res.apply(lambda x: len(x['Name']), axis = 1)
df_res
```

```python
df_res.groupby(['Name', 'Gender'])[['Quantity']].agg(['sum', 'max'])
```

```python
stacked_df = df_res.groupby(['Gender', 'Year']).mean()
stacked_df
```

```python
unstacked_df = stacked_df.unstack('Gender')
unstacked_df
```
